```python
sns.countplot('category',data=info_Data)

plt.xticks(rotation=45);
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(info_Data.corr(),annot=True)
plt.xticks(rotation=90);
```

```python
plt.figure(figsize=(12,8))
sns.regplot(x='price',y='num_subscribers',data=info_Data)
```

```python
info_DataNum = info_Data.loc[:, info_Data.dtypes == np.int64]
print("Data dims : ", info_DataNum.shape) 
info_DataNum.info()   # note that all variables are now int64
```

```python
sns.pairplot(info_Data, hue="is_paid")
```

```python
sns.set(style="darkgrid")
plt.subplots(figsize = (15,8))
ax = sns.barplot(x = "category", 
                 y = "num_subscribers", 
                 data=info_Data, 
                linewidth=5
                )

plt.title("Price Distribution Across Courses offered", fontsize = 25,loc = 'center', pad = 40)
plt.ylabel("num_subscribers", fontsize = 15, )
plt.xlabel("category",fontsize = 15);
plt.xticks(rotation=90);
```

```python
plt.rcParams['figure.figsize'] = (20, 8)
sns.countplot(info_Data['price'], palette = 'rainbow')
plt.title('Distribution of Courses Cost', fontsize = 20)
plt.show()
plt.rcParams['figure.figsize'] = (20, 8)
sns.countplot(info_Data['category'], palette = 'rainbow')
plt.title('Distribution of levels', fontsize = 20)
plt.show()
plt.rcParams['figure.figsize'] = (20, 8)
sns.countplot(info_Data['is_paid'], palette = 'rainbow')
plt.title('Distribution of Paid/Free Courses', fontsize = 20)
plt.show()
plt.xticks(rotation=90);
```
